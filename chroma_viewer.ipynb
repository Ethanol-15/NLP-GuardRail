```python
import chromadb

# Initialize a persistent client if your DB is stored on disk
client = chromadb.PersistentClient(path="./chroma_db")

# Get a specific collection
collection = client.get_collection(name="knowledge_base")

# Retrieve records from the collection
# You can specify IDs, a limit, or an offset for pagination
results = collection.get(limit=100) # Get the first 10 records

# Print the retrieved documents and metadatas
for i in range(len(results['documents'])):
    print(f"Document: {results['documents'][i]}")
    print(f"Metadata: {results['metadatas'][i]}")
```
